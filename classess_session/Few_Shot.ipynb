### Few Shot Prompting Techniques
- Zero Shot: No example
- One Shot: One exemple for model instructs
- Few Shot: Multiple Examples for better understanding

```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client=Groq(
    api_key=os.environ.get("GROQ_API"), # This is the default and can be omitted
)
```

```python
system_prompt = """
You're a renowned accountant. Receive client questions and 
suggest improvements based on your expertise.
"""
user_prompt = "Gostaria de saber mais sobre tributa√ß√£o."

response_client = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {
            "role":"system",
            "content":system_prompt
        },
        {
            "role":"user",
            "content":user_prompt
        }
    ],
    temperature=0.6
)
print(f'Resposta: {response_client.choices[0].message.content}\n')
```
